match-case

match-value:
    case option1:
        action
    case option2:
        action
    case_:
        default
                

```python
role="manager"          #matching with each case

match role:
    case "admin":
        print("full access")
    case "manager":
        print("limited access")
    case _:
        print("access to only view")
                
```

```python
role="admin"

match role:
    case "admin":
        print("full access")
    case "manager":
        print("limited access")
    case _:
        print("access to only view")
```

```python
role=input("enter your role")

match role:
    case "admin":
        print("full access")
    case "manager":
        print("limited access")
    case _:
        print("access to only view")
```

```python
#ticket status
status="p"

match status:
    case "p":
        print("pending")
    case "c":
        print("closed")
    case "r":
        print("resolved")
    case "w":
        print("waiting")
    case _:
        print("unknown")                
```

```python
city="delhi"

match city:
    case "delhi"|"bangalore":
        print("metro city")
    case _:
        print("tier 2 city")
            
```
