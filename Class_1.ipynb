f-string formatting types-
variable {x}
expression {x+y}
precision {x:.2f}
width {x:8}
alignment {x:^8}
padding {x:*^8}
zero padding {x:05}
percentage {x:.2%}
commas {x:,}


```python
print(f"the sum of numbers is {a+b}")   #expression
price=3600.78965
print(f"{price:.2f}")   #precision

x=42
print(f"{x:7}")     #width

p="python"
print(f"|{p:<15}|")     #left alignment
print(f"|{p:>15}|")     #right alignment
print(f"|{p:^15}|")     #centre alignment


print(f"{42:*<6}")      #padding
print(f"{42:*>6}")
print(f"{42:*^6}")



value=3.5
print(f"{value:07.2f}")     #zero padding 


ratio=0.7865
print(f"{ratio:.2%}")          #percentage formatting



salary=125000       #commas (thousand seperator)
print(f"{salary:,}")
print(f"{salary:,.2f}")




```

#TUPLES-
Tuples are ordered, heterogenous, immutable(not changeable),allows duplicates


```python
a=(1,2,3)
print(type(a))


a=(1,2,3,"red",1,2,3,4,5)
print(type(a))

print(a[0])     #indexing
print(a[0:3])   #slicing
```

```python
#tuple - methods are count,index

a=(3,4,3,3,6,5,7,4,3,6,)
print(a.count(3))
print(a.index(6))



t=6,7,8     #tuple packing (packing all given elements)
print(type(t))

a,b,c=t     #tuple unpacking
print(a)
print(b)
print(c)
```

```python
#built in functions that can be used for tuple
a=(3,4,6,7,8,3)
print(len(a))
print(max(a))
print(min(a))
print(sum(a))
```
