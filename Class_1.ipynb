```python
#relationaloperators
a=10
b=5

print(a==b)     #equal to
print(a<b)      #less than
print(a>b)      #greater than
print(a!=b)     #not equal to
print(a<=b)     #less than or equal to
print(a>=b)     #greater than or equal to

```

```python
#logical operators
print((b<a) and (b!=a))
print((a<b) or (b==a))
print(not (a<b)) 
print((a<b) and (b!=a))


print(False and True)
print(False or True)




```

```python
#bitwise operators      #negation -(n+1)(~)


print(5 & 3)    #binary and
print(5 | 3)    #binary or
print(~5)       #negation
print(5 ^ 3)    #binary XOR
print(5<<1)     #left shift (shifting decimal places one by one to the left side)
print(5>>1)     #right shift (shifting decimal places one by one to the right side)

```

f-string formatting types-
variable {x}
expression {x+y}
precision {x:.2f}
width {x:8}
alignment {x:^8}
padding {x:*^8}
zero padding {x:05}
percentage {x:.2%}
commas {x:,}


```python
print(f"the sum of numbers is {a+b}")   #expression
price=3600.78965
print(f"{price:.2f}")   #precision

x=42
print(f"{x:7}")     #width

p="python"
print(f"|{p:<15}|")     #left alignment
print(f"|{p:>15}|")     #right alignment
print(f"|{p:^15}|")     #centre alignment


print(f"{42:*<6}")      #padding
print(f"{42:*>6}")
print(f"{42:*^6}")



value=3.5
print(f"{value:07.2f}")     #zero padding 


ratio=0.7865
print(f"{ratio:.2%}")          #percentage formatting



salary=125000       #commas (thousand seperator)
print(f"{salary:,}")
print(f"{salary:,.2f}")




```
