Objective:
build a script that:

1. Integration - fetch data from API

2. convert + clean data

3. output -> save to csv with timestamp


```python
#step 1 - integration (api call)
import requests
url="https://api.coingecko.com/api/v3/coins/markets"
params={
    "vs_currency":"usd","per_page":5
}
response=requests.get(url,params=params)
```

```python
#step 2 check response
if response.status_code==200:
    data=response.json()
else:
    print("Error:",response.status_code)    
```

```python
#Step 3 - processing (json -> data)
import pandas as pd
df=pd.DataFrame(data)
print(df.head())
```

```python
#Step 4 select required column
df=df[["name","current_price","market_cap"]]
print(df)
```

```python
#step 5 handling missing values
df.fillna(0,inplace=True)
```

```python
#step 6 - create timestamp
from datetime import datetime
timestamp=datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
```

```python
#step 7 - output (save to csv)
import os
os.makedirs("output",exist_ok=True)

filename=f"crypto_data_{timestamp}.csv"
path=os.path.join("output",filename)
df.to_csv(path,index=False)
print("Saved:",path)
```
