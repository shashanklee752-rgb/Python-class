API- Application Programming Interface

It allows different programs to talk to each other
(to get or senf data)

Client(python) -> request -> server(API) -> response -> client

Endpoint
payload 
headers

endpoint is the API url (this is the url where API is accessed)

payload is the data sent. used in query parameters (get) and the body (post)

headers - extra info sent with request 

```python
#requests library
import requests
url="https://api.coingecko.com/api/v3/ping"

response=requests.get(url)

print(response.status_code)
print(response.json())  #java-script-object-notation
```

```python
url="https://api.coingecko.com/api/v3/coins/markets"
q={"vs_currency":"usd","per_page":2}

print(requests.get(url,params=q).json())
```

```python
#post request
#post is sending data to the server to create or updata something

url="https://httpbin.org/post"
data={"name":"Alice"}

response=requests.post(url,json=data)
print(response.json())
```

```python
url="https://reqres.in/api/users"

data={
    "name":"Alice",
    "job":"Data Analyst"
}

headers={"user-Agent":"google chrome"} #to say who you are
response=requests.post(url,json=data,headers=headers)
print(response.status_code)
print(response.json())
```

Response handling - status codes

200- success
404- not found
500- server error

```python
#check before using data
if response.status_code==200:
    data=response.json()
    print(data)
else:
    print("Error:",response.status_code)    
```
