concatenation- pd.concat()

concatenation means combinig dataframes by stacking them

two main ways to stack:-

vertical stacking      axis=0
horizontal stacking    axis=1

```python
import pandas as pd
df1=pd.DataFrame({
    "Customer":["Asha","Bala"],
    "City":["Bangalore","Mumbai"]
})
df1
```

```python
#stack rows
#vertical stackings or adding more rows

pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0).reset_index(drop=True)
```

```python
df3=pd.DataFrame({"Sales":[50000,30000]})

df4=pd.DataFrame({"Quantity":[2,1]})
```

```python
#horizontal stacking or adding more columns
pd.concat([df3,df4],axis=1)
```

```python
#vertical concatenation

import pandas as pd
jan=pd.DataFrame({
    "Customer":["Asha","Sanjana"],
    "City":["Bangalore","Mumbai"],
    "Sales":[5000,7000]
})
jan

feb=pd.DataFrame({
    "Customer":["Chris","Deepak"],
    "City":["Delhi","Kolkata"],
    "Sales":[6500,7200]
})
feb
print()
```

```python
pd.concat([jan,feb],axis=0).reset_index(drop=True)
```

```python
#horizontal stacking
customers=pd.DataFrame({
    "Customers":["Asha","Bala","Chitra"]
})

sales=pd.DataFrame({
    "Sales":[5000,7000,8000]
})

pd.concat([customers,sales],axis=1)
```

```python
df1=pd.DataFrame({
    "customer":["Asha","Bala"],
    "Sales":[5000,6000]
})
df2=pd.DataFrame({
    "Customer":["Chitra","Deepa"],
    "Profit":[9000,10000]
})
pd.concat([df1,df2]).reset_index(drop=True).fillna(0)
```

#merging (combining tables)
#merge is used when data is stored in seperate tables but linked by keys

Syntax-
pd.merge(left_df,right_df,on="column")

```python
customers=pd.DataFrame({
    "CustomerID":[1,2,3],
    "CustomerName":["Asha","Deepa","Chiranth"],
    "City":["Bangalore","Kolkata","Jaipur"]
})
print(customers)
print()
orders=pd.DataFrame({
    "CustomerID":[1,2,4],
    "OrderValue":[5000,7000,6500]
})
print(orders)
```

```python
pd.merge(customers,orders,on="CustomerID")
#this will match rows where customer ID exists in both tables
```

```python
#when column names don't match
df1=pd.DataFrame({
    "custID":[1,2],
    "Name":["Anusha","Rajesh"]
})

df2=pd.DataFrame({
    "CustomerID":[1,2],
    "Sales":[5000,7000]
})
```

```python
pd.merge(df1,df2,left_on="custID",right_on="CustomerID")
```

```python
#to remove extra key column while merging
#you can rename the column before merge
df1=pd.DataFrame({
    "custID":[1,2],
    "Name":["Anusha","Rajesh"]
})

df2=pd.DataFrame({
    "CustomerID":[1,2],
    "Sales":[5000,7000]
})
df2=df2.rename(columns={"CustomerID":"custID"})
pd.merge(df1,df2,left_on="custID")
```

```python
#outer join - all rows from both tables
pd.merge(customers,orders,on="CustomerID",how="outer")
```

```python
#joining through multiple keys
sales=pd.DataFrame({
    "CustomerID":[1,1,2],
    "Product":["Laptop","Mobile","Tablet"],
    "Sales":[5000,3000,7000]
})

prices=pd.DataFrame({
    "CustomerID":[1,2],
    "Product":["Laptop","Tablet"],
    "Discount":[10,15]
})
pd.merge(sales,prices,on=["CustomerID","Product"])
```

```python
#duplicate keys

df1=pd.DataFrame({
    "ID":[1,1],
    "Name":["Asha","Asha"]
})

df2=pd.DataFrame({
    "ID":[1,1],
    "Sales":[5000,6000]
})
pd.merge(df1,df2,on="ID")
#many to many join
```

```python
#suffixes
#when both tables contain same column name
df1=pd.DataFrame({
    "ID":[1,2],
    "Sales":[5000,7000]
})

df2=pd.DataFrame({
    "ID":[1,2],
    "Sales":[3000,4000]
})
pd.merge(df1,df2,on="ID",suffixes=("_store1","_store2"))
```
