pandas is a python libraries used for:

- reading data(csv,excel,etc)
- cleaning data
- analysing data
- preparing data for visualization


```python
import pandas as pd
```

```python
#series vs dataframe
#a series is a single column of data
#it is one dimensional
```

```python
s=pd.Series([1,2,3,4])
print(s)
```

```python
#custom index
s=pd.Series([1,2,3,4],index=['a','b','c','d'])
print(s)
```

```python
s=pd.Series([1.1,2.5,3.7,4.9],index=['a','b','c','d'])
print(s)
```

```python
#dataframe
#a dataframe is a 2D table of data(rows+columns)
data={'name':['Alice','Bob','Charlie','David'],
      'age':[25,30,35,40],
      'city':['NY','LA','Chicago','Houston']}
df=pd.DataFrame(data)
print(df)
print(df.index)
print(df.columns)
print(df.values)
```
