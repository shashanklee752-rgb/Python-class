```python
import pandas as pd
import numpy as np

data = {
    'customer_name': ['Shash', 'Bala', 'Deepa', 'Esha', 'Bala', 'park','bala','bala'],
    'city': ['Bangalore', 'Delhi', 'Delhi','Bangalore', 'Delhi','bangalore','delhi','delhi'],
    'order_value': [5000, 7000, np.nan, 6500, 7000, 6500,7000,7000],
    'Email': ['shash@gmail.com', 'bala@reddif.com', 'bala@gmail.com',
             'esha@gmail.com', 'bala@yahoo.com', 'park@gmail.com','bala@gmail.com','bala@gmail.com']   
}
df = pd.DataFrame(data)
df
```

#Handling duplicate values
duplicates occurs when the same record appears multiple times in a dataset

common reasons :-
data entry errors
merging datasets
system logging errors
duplicates can distort the analysis, for ex
revenue may appear higher than reality


```python
#viewing duplicate rows
df[df.duplicated()]
```

```python
#identify the duplicate values
df.duplicated()
```

```python
#removes completely identical rows
df.drop_duplicates()
```

```python
df1=df.drop_duplicates()
print(df1)
print(df)
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
#using subset
df1=df.drop_duplicates(subset='Email')
df1
#this removes duplictes only based on the email column
```

```python
df1=df.drop_duplicates(subset='city')
df1
```

```python
#keep decides which duplicates record remains
df1=df.drop_duplicates(keep='first')
df1
```

```python
#keep decides which duplicates record remains
df1=df.drop_duplicates(keep='last')
df1
```

```python
#keep decides which duplicates record remains
df1=df.drop_duplicates(keep=False)
df1
#keep parameters accepts only three values-first-last-False
```

```python
#REGEX - regular expressions
#to extract patterns from text

```

common regex patterns

\d               digit
\w               letter or number
^                start of a string
$                end of string 
.                any character
+                one or more
-                zero or more


commonly used with - 
.str.contains()
.str.extract()
.str.replace()
.str.match()


```python
import pandas as pd
data={"Email":["asha123@gmail.com",
               "bala@yahoo.com",
               "shash@gmail.com",
               "chitra_98@company.com"]}
df=pd.DataFrame(data)
```

```python
#extract numbers from text
df["Email"].str.extract(r"(\d+)")
```

```python
#find rows containing numbers
df["Email"].str.contains(r"\d")
```

```python
#extract email domain
df["Email"].str.extract(r"@(\w+)")
```

when we write @(\w+)
()                   capture group
\w                   letter or number
+                    one or more

here match starts at the symbol @ but since @ is outside of the capture group it is not extracted so only what is inside capture group() is extracted 

```python
#find emails ending with gmail.com
df["Email"].str.contains(r"@gmail\.com$")
```

```python
data={"name":["asha","bala","chitra","deepa"]}
df=pd.DataFrame(data)
```

```python
#find names starting with A
df["name"].str.contains(r"^a")
```

```python
#remove numbers from emails
df["Email"].str.replace(r"\d"," ")
```

```python
#regex groups
#chitra98@gmail.com
#we want to extract both name and number in the email seperately
df["Email"].str.extract(r"(\w+)_(\d+)")
```
