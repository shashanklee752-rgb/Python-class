web scraping -
web scraping = extracting data from websites using code instead of copying manually

Request webpage -> get HTML -> parse HTML -> extract data

```python
#web scraping
import requests
from bs4 import BeautifulSoup   #is for getting html requests
url="https://www.google.com/maps/"

response=requests.get(url)
soup=BeautifulSoup(response.text,"html.parser")
print(soup.title.text)
```

```python
import requests
from bs4 import BeautifulSoup   #is for getting html requests
url="https://www.google.com/maps/"

response=requests.get(url)
if response.status_code==200:
    print("Success")
    
soup=BeautifulSoup(response.text,"html.parser")
print(soup.title.text)  #Extract title
```

```python
#extract paragraph

paragraphs=soup.find_all("p")

for p in paragraphs:
    print(p.text)
```

```python
links=soup.find_all("a")

for link in links:
    print(link.get("href"))
```
