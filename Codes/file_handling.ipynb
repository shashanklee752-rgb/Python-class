when you open a file python creates:
1. a connection to a file
2. a file object in memory(RAM)
3. a system resource handle

if you don't close it that connection stays open

if you leave it open
resources get wasted
system may slow down
you may not be able to open a new file

```python
#exception handling(error handling)
#try-except
try:
    num=int(input("enter a number:"))
except ValueError:
    print("invalid input,please enter a number")
```

ValueError - wrong value
Type Error - wrong data type
NameError - variable not defined
ZeroDivisionError - divide by zero error
FileNotFoundError - file missing
IndexError - list index wrong
KeyError - dictionary key missing
AttributeError - method doesn't exist
ImportError - module cannot be imported
Exception - catches almost all errors(not recomended for specfic error handlings)


```python
try:
    print(age)
except NameError:
    print("age is not defined")        
```

```python
try:
    print(10/0)
except ZeroDivisionError:
    print("cannot divide by zero")
```

```python
try:
    open("abc.txt","r")
except FileNotFoundError:
    print("file not found")
```

```python
try:
    numbers=[1,2,3]
    print(numbers[5])
except IndexError:
    print("index out of range")
    
```

```python
try:
    data={"name":"shashank"}
    print(data["age"])
except KeyError:
    print("key not found")
```

```python
#exception is generic,it catches almost all errors
try:
    x=10/0
except Exception:
    print("an error occurred:")
    #not recommended for specific error
    
    
```

```python
#catching multiple errors
try:
    a=int(input())
    b=int(input())
    print(a/b)
except (ValueError,ZeroDivisionError):
    print("invalid input")
```

finally-

finally is a block that always runs whether there is an error or not

basic syntax:
try:
    #risky code
except:
    #handle error
finally:
    #always runs

    we use finally to
1. close the file
2. close database connections
3. release resources
4. cleanup actions    



```python
try:
    print(10/0)
except ZeroDivisionError:
    print("cannot divide by zero")
finally:
    print("done")        
```

```python
#with return statements
def test():
    try:
        return "try block"
    finally:
        print("finally block")
        
print(test())
```
