```python
#return with condition
def check_even(num):
    if num%2==0:
        return "even"
    else:
        return "odd"
print(check_even(3))    
    
```

```python
#function stop when it executes return
def find_first_even(nums):
    for n in nums:
        if n%2==0:
            return n
            print("hey")        #this will not be executed since return has been executed
        return None
print(find_first_even([10,12,13,19]))
print(find_first_even([11,17,13,19]))    
```

```python
#return list
def get_squares(nums):
    result=[]
    for n in nums:
        result.append(n*n)
    return result
print(get_squares([2,5,9]))    
    
```

```python
#return boolean
def is_positive(num):
    return num>0
print(is_positive(-4))
```

```python
#return dictionary
def create_students(name,age):
    return {"name":name,"age":age}
print(create_students("sameeha",22))
```

Docstring-
a string written inside a function to explain what it does

it helps:
other developers
future you
interviewers
documentation tools

it uses triple quotes- """ """


```python
def add(a,b):
    """
    adds two numbers.
    """
    return a+b
add(5,7)

print(add.__doc__)  #to access the docstring written inside
```

```python
#google style docstring(format)
def add(a,b):
    """
    add two numbers.
    
    args:
        a(int):first number
        b(int):second number
    returns:
    int:sum of a and b
    """
    return a+b  
```

```python
#numpy style docstring(format)
def add(a,b):
    """
    add two numbers.
    
    parameters
    ----------
    a : int
        first number.
    b : int
        second number.
        
    returns
    -------
    int
        sum of a and b
    """
    return a+b           
```

```python
# *args- variable positional arguments
# *args collects extra positional arguments into a tuple
#non variable arguments
#sum all numbers
def total(*numbers):
    total_sum=0
    for n in numbers:
        total_sum+=n
    return total_sum

print(total(10,20,40,50,38,47,23.5))
print(total(50,60,45))
    
```

```python
def greet(greeting,*names):
    for name in names:
        print(greeting,name)
        
greet("Good Morning","shankar","ashwin","natille")        
        
```

```python
def show_types(*args):
    for value in args:
        print(value,"->",type(value))
show_types(10,"sam",5.5)        
```

```python
#kwargs
#key word arguments
def show_details(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)
show_details(name="sam",age=22,city="Bangalore")     
        
```

```python
def check_email(**kwargs):
    if "email" in kwargs:
        return "email provided"
    return "email missing"
print(check_email(name="sam",age=22,email="test@gmail.com"))
```

```python
#update default settings
def settings(**red):
    config={"theme":"light","font_size":12,}
    config.update(red)
    return config
print(settings(theme="dark"))
```

```python
#combined example
def example(a,*args,**kwargs):
    print("a:",a)
    print("args:",args)     #it takes only numric values
    print("kwargs:",kwargs) #it takes only keywords
example(10,20,30,name="sam",song="retro")
```

#lambda (anonymous function)

lambda= small function written in one line

syntax:
lambda parameters: expression

no return keyword required

use lambda for-

short logic 
inside map filter
one-time use

```python
#normal function
def square(x):
    return x*x

#lambda function (single parameter)
#lambda function
square=lambda x:x*x
print(square(5))

#lambda function with multible parameters
add=lambda a,b:a+b
print(add(5,3))
```

```python
# check even
is_even=lambda num:num%2==0
print(is_even(3))

#get last character
last_char=lambda s:s[-1]
print(last_char("python"))

#convert to upper case
upper=lambda s:s.upper()
print(upper("hello world"))

#convert to lower case
lower=lambda s:s.lower()
print(lower("HELLO WORLD"))
```

```python
#map()
#applies function to every element
#normal loop
nums=[1,2,3]
squares=[]

for n in nums:
    squares.append(n*n)
print(squares)

#using map()
nums=[1,2,3]
squares=list(map(lambda n:n*n,nums))
print(squares)
```

```python
#add 10 to each
nums=[5,10,15]
result=list(map(lambda n:n+10,nums))
print(result)

#uppercase names
names=["sam","ashwin","natille"]
upper_names=list(map(lambda s:s.upper(),names))
print(upper_names)
```

```python
#filter function
#returns elements that satisfy a condition

#normal loop
nums=[1,2,3,4,5]
even_nums=[]
for n in nums:
    if n%2==0:
        even_nums.append(n)
print(even_nums)

#using filter()
nums=[1,2,3,4,5]
even_nums=list(filter(lambda n:n%2==0,nums))
print(even_nums)
```

```python
#remove empty strings
data=["hello","","world","","python"]
result=list(filter(lambda s:s!="",data))
print(result)
```

```python
#reduce
#used for cummulative calculation
#have to import for this

from functools import reduce
nums=[1,2,3,4]
product=reduce(lambda a,b:a+b,nums)
print(product)
```

```python
#concatenate strings
words=["hello","world","python"]
print(reduce(lambda a,b:a+" "+b,words))
```

```python
#find max
nums=[5,2,9,1]
max_num=reduce(lambda a,b:a if a>b else b,nums)
print(max_num)
```

#scope- LEGB RULE
#scope determines where the variable lives

LEGB stands for

L-Local
E-Enclosing
G-Global
B-Built-in


```python
#local scope
def add(a,b):
    a=5
    b=10
    print(a+b)      #here a and b are local variables, and they exist only inside a function
    
add(5,10)    
```

```python
#global scope
age=24

def show():
    print(age)
    
show()
#here age is the global variable
#function can read global variable
    
```

```python
#enclosing scope
def outer():
    x=10
    def inner():
        print(x)    #inner function can access variable of outer function
    inner()

outer()

#inner function uses variable from the outer function
#here x is enclosing  scope variable
    
```

```python
#built-in scope
k="names"
print(k)
print(len(k))    
```

```python
del len #deleting the inn object variable
print(len(k))
```

```python
#global + local same name
x=100
def show():
    x=50
    print("inside function",x)
    
show()
print("outside function",x)
#local variable does not affect or change global variable
        
```

```python
#using global variable inside a function
total=0
def add_amount(amount):
    global total
    total+=amount
add_amount(3000)
add_amount(500)
print(total)
    
```
