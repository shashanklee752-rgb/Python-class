import matploptlib.pyplot as plt

there are two ways to create plots

1. pyplot (plt.plot)

2. object oriented API (recommended)

fig,ax=plt.subplots()

fig = entire figure or canvas

ax = axes where the chart is drawn


```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
plt.plot(months,sales)

plt.title("Monthly sales trend")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]

fig,ax=plt.subplots()
ax.plot(months,sales)
ax.set_title("Monthly sales trend")
ax.set_xlabel("months")
ax.set_ylabel("sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]

fig,ax=plt.subplots(1,2)
ax[0].plot(months,sales)
ax[1].bar(months,sales)

plt.show()
```

types of plots:
line plot - these are used to show the trend across time or sequence

bar plot - this compares values across categories

scatter plot - this shpws the relationship between two numeric values

```python
#bar plot-

products=["product A","product B","product C","product D"]
sales=[120,180,150,200]
plt.bar(products,sales)
plt.title("Sales by product")
plt.xlabel("products")
plt.ylabel("sales")
plt.show()
```

```python
#scatter plot 
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
advertising=[50,60,70,80,90]

plt.scatter(advertising,sales)
plt.title("Advertising v/s Sales")
plt.xlabel("Advertising")
plt.ylabel("Sales")
```

```python
#plot customisation
#titles
#axis labels
#ticks
#legends
#annotations

#axis label and title
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
profit=[50,60,70,80,90]
plt.plot(months,sales)
plt.title("Monthly Sales Trend")
plt.xlabel("months")
plt.ylabel("Sales")
```

```python
#linestyle and line color with markers
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
profit=[50,60,70,80,90]
plt.plot(months,sales,color="red",linestyle=":",marker="x")
plt.title("Monthly Sales Trend")
plt.xlabel("months")
plt.ylabel("Sales")
#colors = red,blue,green,black
#linestyle = - (solid), --(dashed), :(dotted)
```

```python
#legend
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
profit=[50,60,70,80,90]
plt.plot(months,sales,label="Sales",color="red",linestyle="--",marker="o")
plt.plot(months,profit,label="profit",color="pink",linestyle=":",linewidth=3)
plt.title("Sales v/s Profit Trend")
plt.xlabel("months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
#grid
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
profit=[50,60,70,80,90]
plt.plot(months,sales,label="Sales",color="red",linestyle="--",marker="o")
plt.title("Sales v/s month")
plt.xlabel("months")
plt.ylabel("month")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#xticks and yticks
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
profit=[50,60,70,80,90]
plt.plot(months,sales,label="Sales",color="red")
plt.title("Sales v/s month")
plt.xlabel("months")
plt.ylabel("month")
plt.xticks(rotation=45)
plt.yticks([200,250,300,350])
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may"]
sales=[200,220,250,270,300]
profit=[50,60,70,80,90]
plt.style.use("ggplot")     #using styling themes
plt.plot(months,sales,label="Sales",color="red")
plt.title("Sales v/s month")
plt.xlabel("months")
plt.ylabel("month")
plt.xticks(rotation=45)
plt.yticks([200,250,300,350])
plt.show()
```
