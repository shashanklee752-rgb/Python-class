```python
import pandas as pd
df=pd.DataFrame({
    "hours":[2,4,6,8],
    "score":[50,65,78,90]
})
```

```python
#matplotlib version
import matplotlib.pyplot as plt
plt.scatter(df["hours"],df["score"])
plt.xlabel("Hours studied")
plt.ylabel("Score")
plt.title("study v/s score")
plt.show()
```

```python
#seaborn version
#scatterplot example
import seaborn as sns
sns.scatterplot(data=df,x="hours",y="score")
plt.show()
```

```python
#Seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "student": ["A","B","C","D","E","F","G","H"],
    "hours_studied": [2,4,6,3,5,7,8,4],
    "exam_score": [50,65,78,60,72,85,90,68],
    "gender": ["F","M","F","M","F","M","F","M"],
    "course": ["Python","Python","SQL","SQL","Python","SQL","Python","SQL"]
}

df = pd.DataFrame(data)

print(df)
```

```python
#scatterplot with category coloring
import seaborn as sns
sns.scatterplot(data=df,x="hours_studied",y="exam_score",hue="gender")
plt.show()
```

```python
#barplot
#average score by course
sns.barplot(data=df,x="course",y="exam_score")
plt.show()
```

```python
#histogram
#distribution of exam scores
sns.histplot(data=df,x="exam_score",bins=5)
plt.show()
```

```python
#with KDE  - Kernel Density Estimation
#it estimates the shape of a data distribution without relying on fixed bins

sns.histplot(data=df,x="exam_score",kde=True)
plt.show()
```

```python
#KDEplot
sns.kdeplot(data=df,x="hours_studied")
plt.show()
```

```python
#boxplot
#score distribution by course
sns.boxplot(data=df,x="course",y="exam_score")
plt.show()
```

```python
#violinplot
#score distribution by gender
#combination of boxplot and density plot
sns.violinplot(data=df,x="gender",y="exam_score")
plt.show()
#shape of distribution of data
#density of values
```
