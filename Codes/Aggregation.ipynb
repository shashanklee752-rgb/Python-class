Aggregation means summarizing the data:
example:

total sales
average revenue
frequency counts

```python
#GroupBy
import pandas as pd

data = {
    "Customer": ["Asha","Bala","Chitra","Deepa","Esha","Farhan","Gita","Hari"],
    "City": ["Bangalore","Delhi","Mumbai","Delhi","Bangalore","Mumbai","Delhi","Bangalore"],
    "Product": ["Laptop","Mobile","Laptop","Tablet","Mobile","Laptop","Tablet","Mobile"],
    "Sales": [50000,30000,45000,20000,32000,47000,22000,31000],
    "Quantity": [2,1,1,3,2,1,2,1]
}

df = pd.DataFrame(data)
df
```

```python
#Group by works with split,apply,combine logic
#total sales by city
df.groupby("City")["Sales"].sum()
```

```python
df.groupby("City")["Sales"].count()
```

```python
#sales by city and product
df.groupby(["City","Product"])["Sales"].sum()
```

```python
df.groupby(["Product","City"])["Sales"].sum()
```

```python
#multiple aggregations
#using .agg()

df.groupby("City").agg({
    "Sales":"sum",
    "Quantity":"mean"
})
```

```python
#multiple aggregations on the same column
df.groupby("City")["Sales"].agg(["sum","mean","])
```

```python
#groupby size vs count

df.groupby("City").size()   #counts number of rows per group

df.groupby("City")["Sales"].count() #counts null values in sales column
```

```python
#pivot tables
#pivot tables reshape data like excel pivot tables
pd.pivot_table(df,values="Sales",
               index="City",
               columns="Product",
               aggfunc="sum")
```

```python
#pivot table with multiple aggregation
pd.pivot_table(df,values="Sales",
               index="City",
               columns="Product",
               aggfunc=["sum","mean"])
```

```python
#crosstab
#crosstab counts occurances it's also called frequency table
pd.crosstab(df["City"],df["Product"])
#this gives us the product popularity in each city
```

```python
#pivot tables with multiple values
pd.pivot_table(df,values=["Sales","Quantity"],
               index="City",
               columns="Product",
               aggfunc="sum")
```

```python
#pivot tables with fill values
pd.pivot_table(df,values="Sales",
               index="City",
               columns="Product",
               aggfunc="sum",
               fill_value=0)
#replace NaN with zero
```
