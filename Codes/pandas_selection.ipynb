#pandas selection
label based selection using .loc
this is used when selecting by row labels and column names

syntax:
df.loc[row_label,column_label]

#label- based selection
1. .loc
#used when selecting by row labels and column names



```python
import pandas as pd
df=pd.read_csv("North America Restaurants (1).csv")
df.head()
```

```python
import pandas as pd
df.loc[0,"name"]  #to select single value
```

```python
df.loc[0:2] #multiple rows
#.loc includes end index
```

```python
df.loc[:,["name","city"]]   #to identify specific column data
```

#iloc
poisition based selection
used when selecting by position


syntax:
df.iloc[row_index,column_index]


```python
df.iloc[0,1]  #to select single value using iloc
```

```python
df.head()
```

```python
df.iloc[0:3,0:2]    #to select multiple rows and columns using iloc
#iloc excludes end index
```

```python
#boolean masking (filtering)
#filters rows based on condition
df[df["weighted_rating_value"]>2.5] #single condition
```

```python
#for multiple condition 
df[(df["weighted_rating_value"]>2.5) & (df["weighted_rating_value"]<3.5)]  #to filter rows based on multiple conditions
```

```python
#or condition
df[(df["weighted_rating_value"]>4.5) | (df["pickup_enabled"]==False)]
```

```python
import pandas as pd
```

```python
df[~(df['country'] == 'CA')]    #to filter rows where country is not CA
```

```python
#query method
#similar to sql method
df.query('weighted_rating_value>3.6')   #to filter rows based on multiple conditions using query method
```

```python
df.iloc[1:2]    #to select specific row using iloc
```
