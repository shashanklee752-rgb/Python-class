```python
#defining a function without parameter
def greet():
    print("hello user hope you are doing great")
```

```python
#calling a function without argument
greet()
```

```python
#defining a function with parameter
def greet1(name):
    print("Hello",name)
```

```python
#calling function with argument
greet1("rahul")
```
