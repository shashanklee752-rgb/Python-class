REGEX-  (regular expression)


a way to search,extract,validate, or replace text using patterns

examples-
find phone numbers
find emails
remove symbols
split messy data

python uses the re module for this
"import re"

most used functions
re.search()
re.findall()
re.sub()
re.split()
re.match()

REGEX SYMBOLS-
PATTERN                                     MEANING
\d                                          digit
\D                                          non-digit
\w                                          letter/number
\W                                          special character
\s                                          space


.                                           any character
+                                           one or more
*                                           zero or more
^                                           start
$                                           end
[abc]                                       any of a/b/c





```python
#re.findall()       (returns all matches) start with r for regex function

import re
text="order 123,amount 450"

nums=re.findall(r"\d+",text)
print(nums)

nums=re.findall(r"\d*",text)
print(nums)

nums=re.findall(r"\d^",text)        #digit
print(nums)

nums=re.findall(r"\D+",text)        #non digit
print(nums)

nums=re.findall(r"\w+",text)        #letters or numbers
print(nums)
```

```python
#re.sub     (replaces using patterns)

phone="+91-987455 3548"
clean_phone=re.sub(r"\D","",phone)
print(clean_phone)


text="hello@#world"     #special characters
re.sub(r"\W"," ",text)


text="data   science"
re.sub(r"\s+"," ",text)         #space characters
```

```python
#re.split()     (for multiple delimiters)

text="python,SQl | excel"
print(re.split(r"[,|]",text))

date="22-12/2025"
print(re.split(r"[-/]",date))




```

```python
#re.search()        checks if it exists

text="my id is 123"
result=re.search(r"\d+",text)
print(result)
```

```python
#re.match() to find the match at the position 1
print(re.match(r"\d","1abc"))
print(re.match(r"\d","abc1"))
```

```python
#extract email username

email="john@gmail.com"
re.findall(r"^\w+",email)
email1="john.123@gmail.com"
re.findall(r"^\w+",email1)
print(email1)
```

```python
#validate pincode (must be 6 digits)

code="560062"
re.match(r"^\d{6}$",code)
bool(re.match(r"^\d{6}$",code))
re.findall(r"^\d{6}$",code)

#since we are validating we need a true or false result hence bool function is used
```
