loops
loops are used to execute the same block of code multiple times without writing the code repeatedly

there are two types of loop
FOR loop
WHILE loop

FOR loop-
a for loop runs once for every item in a collection
collection=list,string,dictionary,etc

syntax-
for item in iterable:
    action


```python
students=["raj","sam","max","rekha"]
for name in students:
    print(name)
```

while loop-
syntax-

   action


```python
#counting example
i=1
while i<=5:
    print(i)
    i+=1
```

```python
#stop at the first failure
marks=[70,80,30,90]
for m in marks:
    if m<40:
        print("fail found")
        break               #break is used to stop loop at a certain point
                            #here break is used to stop at the first failure
                            
```

```python
#continue statement
values=["rahul","","amit"]
for v in values:
    if not v:
        continue
    print(v)
    #empty string is skipped
    #continue is used to skip current iteration and go to the next
    
```

```python
#stop when value found      #here i is index position
#stop the loop when 20 is found
nums=[10,11,12,20,30,40]
i=0
while i<len(nums):
    if nums[i]==20:
        print("value found")
        break
    print("number found is:",nums[i])
    i=i+1
```

```python
#stop when value found
#stop the loop when 20 is found
nums=[10,11,12,20,30,40]
i=0
while i<len(nums):
    if nums[i]==20:
        print("value found")
        break
    i=i+1
```

```python
#skip even numbers between 1 to 10
i=0
while i<=10:
    if i%2==0:
        i=i+1
        continue
    print(i)
    i=i+1
```

```python
#skip odd numbers between 1 to 10
i=0
while i<10:
    i+=1
    if i%2!=0:  
        continue   
    print(i)
```

```python
#find max in a list
nums=[10,45,20,99,30]
max_val=nums[0]
i=0
while i<len(nums):
    if nums[i]>max_val:
        max_val=nums[i]
    i=i+1
print(max_val)        
```

```python
#print square upto 10
i=1
while i<=10:
    square=i**2 
    print(f"Number:{i}, Square:{square}")
    i+=1
```

```python
#print largest even number
nums=[11,22,33,44,66]
largest_even=-1
for n in nums:
    if n%2==0 and n>largest_even:
        largest_even=n
print(largest_even)        
```

```python
#print largest even number
nums=[11,22,33,44,66]
i=0
max_val=0
while i<len(nums):
    if nums[i]%2==0 and nums[i]>max_val:
        max_val=nums[i]
    i=i+1
print("largest even number is:",max_val)        
```
